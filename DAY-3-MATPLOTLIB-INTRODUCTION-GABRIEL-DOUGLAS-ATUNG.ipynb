# MATPLOTLIB
## 1. What is Matplotlib?
- Matplotlib is a python plotting libary that is used for making various visualizations while carrying out data analysis.
- With matplotlib, you can plot different graphs such as line chart, bar chart, pie chart, scattered plot etc

```python
# To begin working on matplotlib, we need to import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
```

```python
# We also need to import numerical values to use matplot, we are importing NumPy
import numpy as np
```

#### Plotting of graphs
Note that, the horizontal axis is called the x-axis while the vertical axis is called the y-axis.

We will plot a graph of x-axis against y-axis

```python
x = np.array([0,6])
y = np.array([0,80])
```

```python
# To make a simple line plot, we have a function called plot as plt
plt.plot(x,y)
# To show the plot, we have to state plt.show()
plt.show()
```

#### Essential use of matplotlib
- Remember always that matplotlib will be use to represent graphical illustrations with sole usage of
     i. plt.plot() and
     ii. plt.show()
  (i & ii) functions will be used to plot graphs while using matplotlib

## Different Types of Matplotlib
### Types of Plot
There are various types of plots in matplotlib namely:
 - Line chart
 - Bar chart
 - Scatter plot
 - Pie Chart
 - Histogram etc.

### a. Line Chart
In line chart, we plot the graph in various ways, it can be line, it can be dotted, it can also be dashed as shown in the various graphs below.

```python
# Let's plot a graph of students score for a game
score = np.array([6,5,14,3,15,10])
plt.plot(score)
# To show only the graph on the screen
plt.show

#LINE GRAPH
```

```python
# For dotted graph
score2 = np.array([8, 12, 3, 14, 5, 6])
plt.plot(score2,linestyle = 'dotted')
plt.show()
```

```python
# For dashed graph
score3 = np.array([6,9,4,15,6,14,3])
plt.plot(score3, linestyle = 'dashed')
plt.show()
```

### b. Bar Chart
Bar chart plotting requires two different types of data in which one will be categorical and the other will be numerical. Also, one will be plotted on y-axis while the other will be plotted on he x-axis

```python
# On the x-axis we can write
x = np.array(["X", "Y", "Z", "W"])
y = np.array([25,10,27,14])
# To plot a barchar, we write plt.bar()
plt.bar(x,y)
plt.show()
```

### c. Scatter Plot
In scatter plot, we work with both numerical features in both the x and the y axis

```python
x1 = np.array([20,15,11,7,8,10])
y1 = np.array([2,7,15,8,10,12])
```

```python
#To plot a scatered plot we write plt.scatter()
plt.scatter(x1,y1)
plt.show()
```

### d. Pie Chart
For Pie chart, we only enter single line numerical values


```python
p = np.array([25,30,45,18,36,27])
# To plot a pie chart we write plt.pie()
plt.pie(p)
plt.show()
```

### e. Histogram
To work with histogram, we can generate random values in numpy as

```python
h = np.random.normal(170,10,250)
```

```python
# To plot a histogram, we write as plt.hist()
plt.hist(h)
plt.show()
```
